Класс Stack с методами push, pop, свойством size и красивым __repr__

```python
class Stack:
    def __init__(self):
        self.arr = []
        self.size = 0
    def push(self, num):
        self.size += 1
        self.arr.append(num)
    def pop(self):
        self.size -= 1
        return self.arr.pop()
    @property
    def Size(self) -> int:
        return self.size
    def __repr__(self):
        return(f"Массив элементов: {self.arr} размером {self.size} элементва/ов")
arr = Stack()
arr.push(1)
arr.push(2)
arr.push(3)
arr.push(4)
print(repr(arr))
arr.pop()
arr.push(5)
print(repr(arr))
```

BankAccount и подклассы SavingAccount и CreditAccount с разной логикой комоссии

```python
class BankAccount:
    def __init__(self, owner: str, balance: float = 0.0):
        self.owner = owner
        self._balance = balance
    @property
    def balance(self) -> float:
        return self._balance
    def deposit(self, amount: float) -> None:
        if amount <= 0:
            raise ValueError("Сумма должна быть > 0")
        self._balance += amount
    def withdraw(self, amount: float) -> None:
        if amount <= 0:
            raise ValueError("Сумма должна быть > 0")
        if amount > self._balance:
            raise ValueError("Недостаточно средств")
        self._balance -= amount
class SavingAccount(BankAccount):
    def withdraw(self, amount: float) -> None:
        if amount > 500:
            comissia = amount + (amount * 0.07)
            super().withdraw(comissia)
        else:
            comissia = amount
            super().withdraw(comissia)
class CreditAccount(BankAccount):
    def withdraw(self, amount: float) -> None:
        if amount > 1000:
            comissia = amount + (amount * 0.11)
            super().withdraw(comissia)
        else:
            comissia = amount
            super().withdraw(comissia)
        if comissia > 2000:
            raise ValueError("Превышает лимит снятия средств")
acc1 = SavingAccount("Jane", 1200)
acc1.deposit(100)
acc1.withdraw(600)
print("Баланс: ", acc1.balance)
acc2 = CreditAccount("Jane", 3000)
acc2.withdraw(1200)
print("Баланс: ", acc2.balance)
```

Абстрактный класс Notifier и реализация EmailNotifier/SMSNotifier

```python
from abc import ABC, abstractmethod
class Notifier(ABC):
    @abstractmethod
    def EmailNotifier(self, notifier: str) -> str: ...
class Message(Notifier):
    def __init__(self):
        self.send = "Пользователь455897"
    def EmailNotifier(self, notifier):
        return f"{self.send}: {notifier}"
message = Message()
stroke = message.EmailNotifier("Привет")
print(stroke)
```

Миксин JsonSerializable с методами to_json/from_json

```python
import json
class JsonSerializable:
    def to_json(self):
        person = {'name': self.name, 'age': self.age, 'gender': self.gender}
        return json.dumps(person, ensure_ascii=False)   
    def from_json(self, jsn: str):
        return json.loads(jsn)
class Person(JsonSerializable):
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
p = Person("Аня", 17, "Женский")
jsn = p.to_json()
print(jsn)
not_jsn = p.from_json(jsn)
print(not_jsn)
```

В проект с фигурами добавить Triangle и проверку сторон

```python
from math import sqrt
from abc import ABC, abstractmethod
from math import pi
from typing import Iterable

class Shape(ABC):
    @abstractmethod
    def area(self) -> float: ...
    @abstractmethod
    def perimeter(self) -> float: ...
    def describe(self) -> str:
        return f"{self.__class__.__name__}(S={self.area():.2f}, P={self.perimeter():.2f})"

class Circle(Shape):
    def __init__(self, r: float):
        self.r = r
    def area(self) -> float:
        return pi * self.r ** 2
    def perimeter(self) -> float:
        return 2 * pi * self.r

class Rectangle(Shape):
    def __init__(self, a: float, b: float):
        self.a, self.b = a, b
    def area(self) -> float:
        return self.a * self.b
    def perimeter(self) -> float:
        return 2 * (self.a + self.b)

class Triangle(Shape):
    def __init__(self, a: float, b: float, c: float):
        self.a, self.b, self.c = a, b, c
        if a + b <= c or a + c <= b or b + c <= a:
            raise ValueError("Такого треугольника не существует")
    def area(self) -> float:
        p = (self.a + self.b + self.c)/2
        return sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))
    def perimeter(self):
        return self.a + self.b + self.c
    
def total_area(shapes: Iterable[Shape]) -> float:
    return sum(s.area() for s in shapes)

def render_report(shapes: Iterable[Shape]) -> None:
    for s in shapes:
        print(s.describe())
    print("Суммарная площадь:", total_area(shapes))

shapes = [Circle(3), Rectangle(2, 5), Triangle(3, 4, 5)]
render_report(shapes)
```
