1. Работа со строками

```python
print("Введите строку: ")
ans = input()
print('Длина строки: ', len(ans))
print('Строка в верхнем регистре: ', ans.upper())
print('Строка в нижнем регистре: ', ans.lower())
if (len(ans) < 3):
    print("Длины строки недостаточно")
else:
    print('Первые три сивола: ', ans[:3])
    print('Последние три сивола', ans[3:])
```

2. Преобразование типов

```python
num1 = str(input('Введите первое число: '))
num2 = str(input('Введите второе число: '))
int_num1 = int(num1)
int_num2 = int(num2)
print('Сумма чисел: ', int_num1 + int_num2)
print('Разность чисел: ', int_num1 - int_num2)
print('Произведение чисел: ', int_num1 * int_num2)
```

3. Поиск пошаблону

```python
import re
text = "Сегодня в 12:30 мы встрчаемся с компанией из 5-ти человек"
pattern = r'\d+'
print('Найденный цифры: ', re.findall(pattern, text))
```

4. Проверка номера телефона

```python
import re
def NumberCheck(number: str) -> bool:
    pattern1 = r'^\+7-\d{3}-\d{3}-\d{2}-\d{2}$'
    pattern2 = r'^8\(\d{3}\)\d{3}-\d{2}-\d{2}$'
    return bool(re.match(pattern1, number)) or bool(re.match(pattern2, number))
tests = ['43439-433', '+7-438-548-33-79', '32-433-433-43-82', '8(439)129-53-76', '+7-342-765-65-21', '+7-43-654-76-34']
for i in tests:
    print(i, '->', NumberCheck(i))
```

5. Валидация пароля

```python
import re
def PasswordCheck(password: str) -> bool:
    pattern = r'^(?=.*[a-z])(?=.*[A-Z])(?=.*\d).{8}$'
    return bool(re.match(pattern, password))
tests = ['fg2Gj', 'Fadk6fds', '34_fdsd2', 'Sort2005', '43Hdf', 'door2011']
for i in tests:
    print(i, '->', PasswordCheck(i))
```
