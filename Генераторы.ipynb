```python
def even(a, b):
    for i in range(a - 1, b):
        if i % 2 == 0:
            yield i
print(list(even(3, 12)))
```

```python
nums = list(range(1, 11))
result = (list(map(lambda x: x*x , filter(lambda x: x % 2 != 0, nums))))
print(result)
```

```python
from itertools import chain
def flatten(iterable):
    return chain.from_iterable(iterable)
print(list(flatten([[1,2],[3],[4,5,6]])))
```

```python
from collections import Counter
text = "Python, python! Генераторы и lambda: быстрее, компактнее и удобнее. Python? Да, Python."
N = 3
def normalize(s):
    for i in s:
        if i.isalnum() or i.isspace():
            yield i.lower()
def words(s):
    strok = ''.join(s)
    for i in strok.split():
        yield i
norm = normalize(text)
word = words(norm)
count = Counter(word)
top = count.most_common(N)
print(top)
```
