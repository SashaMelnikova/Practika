```python
#Импорт библиотеки
import math
#Создание класса
class Drobi:
    #Инициализация атрибутов
    def __init__(self):
        self.chis1 = 0
        self.chis2 = 0
        self.znam1 = 1
        self.znam2 = 1
    #Ввод дробей
    def enter(self):
        self.chis1 = int(input("Введите числитель первой дроби:"))
        self.chis2 = int(input("Введите числитель второй дроби:"))
        self.znam1 = int(input("Введите знаменатель первой дроби:"))
        self.znam2 = int(input("Введите знаменатель второй дроби:"))
    #Вычисление общего знаменателя
    def nok(self, a, b):
        return abs(a * b) / math.gcd(a, b)
    #Сложение дробей
    def summa(self):
        nok = self.nok(self.znam1, self.znam2)
        nchis1 = self.chis1 * (nok / self.znam1)
        nchis2 = self.chis2 * (nok / self.znam2)
        res = nchis1 + nchis2
        print(f"Сумма: {res}/{nok}")
    #Вычитание дробей
    def minus(self):
        nok = self.nok(self.znam1, self.znam2)
        nchis1 = self.chis1 * (nok / self.znam1)
        nchis2 = self.chis2 * (nok / self.znam2)
        res = nchis1 - nchis2
        print(f"Разность: {res}/{nok}")
    #Умножение дробей
    def umn(self):
        res1 = self.chis1 * self.chis2
        res2 = self.znam1 * self.znam2
        print(f"Произведение: {res1}/{res2}")
    #Деление дробей
    def delet(self):
        res1 = self.chis1 * self.znam2
        res2 = self.znam1 * self.chis2
        print(f"Частное: {res1}/{res2}")
    #Выбор действия с дробями
    def choise(self):
        ans = input("Введите, какую операцию вы хотите сделать:")
        match(ans):
            case('+'):
                self.summa()
            case('-'):
                self.minus()
            case('*'):
                self.umn()
            case('/'):
                self.delet()
#Создание объекта класса
drob = Drobi()
drob.enter()
drob.choise()
```
