Функция power(base, exp), которая возводит число base в степень exp

```python
def power(base, exp):
    return base ** exp
power_example = power(2,5)
print(power_example)
```

Функция apply_to_list(func, items), которая применяет func ко всем элементам списка и возвращает новый список

```python
def apply_to_list(func, items = []):
    items = [func(i) for i in items]
    return items
def multiplication(x):
    return x * 2
item = [1, 2, 3, 4, 5]
result = apply_to_list(multiplication, item)
print(result)
```

функция-фабрика make_adder(n), которая возвращает функцию, добавляющую число n

```python
def make_adder(n):
    def adder(x):
        return x + n
    return adder
add = make_adder(10)
print(add(15))
```

декоратор timeit, который измеряет время выполнения функции.

```python
import time
def timeit(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"Функция {func.__name__} выполнялась {end - start:.6f} секунд")
        return result
    return wrapper
@timeit
def power(base, exp):
    return base ** exp
power_example = power(2,5)
print(f"Результат: {power_example}")
```

*args, **kwargs и декоратор в одной задаче.

```python
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"Функция принимает в себя значения {args} и {kwargs}")
        result = func(*args, **kwargs)
        print(f"Функция {func.__name__} вернула {result}")
        return result
    return wrapper
@logger
def power(base, exp):
     return base ** exp
power_example = power(2, exp = 5)
```
